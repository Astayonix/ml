```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
google_data = pd.read_csv('datasets/stocks_goog_1yr.csv', index_col=0)
```

```python
google_data
```

```python
google_data.Date = pd.to_datetime(google_data.Date)
```

```python
google_data.dtypes
```

```python
x = np.array(google_data.Date)
```

```python
y = google_data.Close
```

```python
plt.scatter(x, y)
plt.xlabel('Date')
plt.ylabel('Close (in USD)')
plt.title('Daily Close Value of GOOG')
plt.show()
```
