<h2> The average number of active vehicles total </h2>

```python
import pandas as pd
from pprint import pprint

uber_data = pd.read_csv('datasets/uber_Uber-Jan-Feb-FOIL.csv') #reading in the CSV file
uber_data.rename(columns = {'dispatching_base_number':'Dispatch_Base_No', 'date':'Date', 'active_vehicles':'Active_Cars', 'trips':'Trips'}, inplace=True) #renaming the columns
uber_data.Date = pd.to_datetime(uber_data.Date) #converting the Date column to date-time objects instead of strings

avg_cars = sum(uber_data.Active_Cars) / len(uber_data.Active_Cars)
print avg_cars #shows the total average number of active vehicles
```

<h2> The average number of trips per day </h2>

```python
date_dict = {} #empty date: avg num trips dict
date_set = set() #empty set for unique dates

for date in uber_data.Date:
    date_set.add(date)

for date in date_set:
    avg_trips_per_day = uber_data.Trips[uber_data.Date == date].mean()
    date_dict[date] = avg_trips_per_day

pprint(date_dict) #pretty prints the average number of trips per day
```

<h2> The average number of active vehicles per dispatching base per day </h2>

```python
dispatch_set = set() #empty set for unique dispatch bases
date_set = set() #empty set for unique dates

date_set = set(uber_data.Date) #coverts the date column into a set of unique dates

dispatch_set = set(uber_data.Dispatch_Base_No) #converts the dispatch base column into a set of unqiue dispatch bases

for date in date_set:
    inner_dict[date] = {} #creates the empty inner dictionary {date: dispatch base no}
    for base in dispatch_set:
        avg_no_active_cars = uber_data.Active_Cars[(uber_data.Dispatch_Base_No == base) & (uber_data.Date == date)].mean() #creates the avg num of active cars per date per dispatch base
        inner_dict[date][base] = avg_no_active_cars #fills up the embedded dictionary {date: {dispatch base no: avg number active cars}}

pprint(inner_dict) #pretty prints the average number of active vehilces per dispatching base per day
```
